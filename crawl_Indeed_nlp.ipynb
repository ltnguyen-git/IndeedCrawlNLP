```python
driver = webdriver.Chrome()
url = 'https://vn.indeed.com/'
driver.get(url)
sleep(3)
```

```python
key_words = 'data'
input_field = driver.find_element('xpath','/html/body/div[2]/div[1]/div/span/div[3]/div[1]/div/div/div/div/form/div/div[1]/div/div[1]/div/div[2]/input')
input_field.send_keys(key_words)

location_field = driver.find_element('xpath','/html/body/div[2]/div[1]/div/span/div[3]/div[1]/div/div/div/div/form/div/div[2]/div/div[1]/div/div[2]/input')
location_field.send_keys('Ho Chi Minh City')

find_job_button = driver.find_element('xpath','//*[@id="jobsearch"]/button')
find_job_button.click()
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}
cards = []
arr_link_page = []
base_links = 'https://vn.indeed.com/viewjob'
for i in range(0,100,10):
    link = f'https://vn.indeed.com/jobs?q=data&l=Ho+Chi+Minh+City&start={i}'
    arr_link_page.append(link)


arr_link_page

```

```python
t=[]

for job in arr_link_page:
    r = requests.get(job, headers)
    driver.get(job)
    soup = BeautifulSoup(driver.page_source)
    cards = soup.find_all('a','jcs-JobTitle css-jspxzf eu4oa1w0')
    sleep(1)
    t.append(cards)
```

```python
t
```

```python
link_job= []
for link in t:
    for tag_a in link:
        link_job.append('https://vn.indeed.com' + tag_a.get('href'))
```

```python
link_job
```

```python
job_description =[]
title = []
company = []
website_company = []
for job in link_job:
    driver.get(job)
    sleep(2)
    div_element = driver.find_element('xpath','//div[@class="jobsearch-jobDescriptionText"]')

    # Extract all the text in the div element, including text in nested p tags
    text = div_element.text.replace('\n',' ')
    job_description.append(text)

    t = driver.find_element('xpath','//h1[@class="icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title"]')
    title.append(t.text)
    # try:
    c = driver.find_element('xpath','//div[@class="jobsearch-CompanyInfoContainer"]')
    # except:
    #     try:
    #         c = driver.find_element('xpath','//div[@class="jobsearch-CompanyInfoWithoutHeaderImage jobsearch-CompanyInfoEji"]')
    #     except:
    #         c = driver.find_element('xpath','//div[@class="jobsearch-CompanyInfoWithoutHeaderImage"]')

    
    # Find all the a tags inside the div element
    a_tags = c.find_elements('xpath','.//a')

# Extract the href attribute from each a tag
    href_list = [a.get_attribute('href') for a in a_tags]
    text_list = [a.text for a in a_tags]
    company.append(text_list)
    website_company.append(href_list)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords



stop_words = stopwords.words('english')

with open('vietnamese-stopwords.txt', 'r', encoding='utf-8') as file:
    stop_words_vietnamese = file.read().splitlines()

print(stop_words)
print(stop_words_vietnamese)
```

```python
from nltk import word_tokenize
from nltk.tokenize import word_tokenize
import nltk
# nltk.download('punkt')
import pandas as pd

# Assuming you have a pandas DataFrame called `df` with a 'job_description' column
final_list = []
for index, row in df.iterrows():
    t = word_tokenize(row['job_description'])
    for w in t:
        if w.lower() not in stop_words:
            if w.lower() not in stop_words_vietnamese:
                if len(w) > 4:
                    final_list.append(w)

print(final_list)
```

```python
from collections import Counter
import pprint
counter_words = Counter(final_list)

pprint.pprint(counter_words.most_common(40))
```
